# Nova secção

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv ('/content/dados_rh.csv')
print (df)
```

```
df.head ()
```

```
# 35 colunas e 1470 linhas.
df.shape
```

```
# nenhuma célula nula
df.isnull().sum()
```

```
# tipos de dados: 26 int, 9 object
df.info()
```

```
df2 = df.groupby('Department')['EmployeeNumber'].count().reset_index()
```

```
#Numero de funcionarios por setor
# RH: 63 R&D: 961 SALES: 446
plt.bar(df2['Department'], df2['EmployeeNumber'])
```

```
df3 = df.groupby('Education')['EmployeeNumber'].count().reset_index()
```

```
df3.plot(kind ='line', subplots= True, marker = '^')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Quantidade de Funcionários')
plt.title('Quantidade de Funcionários por Nível de Escolaridade')
```

```
df4 = df.groupby('Attrition')['EmployeeNumber'].count()
```

```
df4.plot(kind ='pie', subplots= True)
```
